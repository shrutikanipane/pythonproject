```python
base_url="https://www.themoviedb.org/movie?page="
movie_url_lst=[]
```

```python
for val in range(1,26):
    movie_url_lst.append(base_url+str(val))
```

```python
import requests
from bs4 import BeautifulSoup
import re
movie_lst=[]
for url in movie_url_lst:
    #url = 'https://www.themoviedb.org'

    header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}

    source_data = requests.get(url, headers = header).text

    soup_data=BeautifulSoup(source_data,"lxml")

    all_divs=soup_data.find_all("div",class_="card style_1")

    
 
    for item in all_divs:#doubt
            movieurl=item.find("a")["href"]
            commonurl="https://www.themoviedb.org"

            url1=commonurl+movieurl
            #print(url1)
            header1 = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}

            #source_data1 = requests.get(url1, headers = header1)
            source_data1 = requests.get(url1, headers = header1).text
            soup_data1=BeautifulSoup(source_data1,"lxml")


            #first_div=soup_data1.find("section",class_="header poster")
            #first_div2=soup_data1.find("section",class_="header poster no_backdrop")
            
            
            h2=soup_data1.find("h2")
            name=h2.find("a").text
            #print(name)
        
        
            if soup_data1.find("span",class_="release")==None:
                releasedate="empty"
            else:    
                releasedate=soup_data1.find("span",class_="release").text
                releasedate=releasedate.strip()
                releasedate=releasedate.split(" ")[0] 
                #print(releasedate)
                
        
            if soup_data1.find("span",class_="genres")==None:
                genres="empty"
            else:    
                genres=soup_data1.find("span",class_="genres").text
                genres=genres.strip()
                #print(genres)
            
            
            if soup_data1.find("span",class_="runtime")==None:
                runtime="empty"
            else:
                runtime=soup_data1.find("span",class_="runtime").text
                runtime=runtime.strip()
                #print(runtime)
            #runtime=first_div.find("span",class_="runtime").text
            #runtime=runtime.strip()
            #print(runtime)
       

            rating=soup_data1.find("div",class_="user_score_chart")
            rating=rating["data-percent"]
            #print(rating)
            
            if soup_data1.find("ol",class_="people no_image")==None:
                ol="empty"
            #div1=first_div.find("div",class_="header_info")
            else:
                ol=soup_data1.find("ol",class_="people no_image")
                #ol=ol.strip()
                ol_li=ol.find_all("li",class_="profile")
                for item in ol_li:
                    p=item.find("p",class_="character").text
                    if p.count("Director")==1:
                        director=item.find("a").text  
                   

            movie_info={
                "name":name,
                "url":url1,
                "releasedate":releasedate,
                "genres":genres,
                "runtime":runtime,
                "rating":rating,
                "director":director  
            }
            movie_lst.append(movie_info)
```

```python
import pandas as pd
df=pd.DataFrame(movie_lst)
```
